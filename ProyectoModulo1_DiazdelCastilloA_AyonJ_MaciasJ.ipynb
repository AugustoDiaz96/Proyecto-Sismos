### Objetivos.

> #### 1.1 Objetivo general.
>  - Simular un sismo para observar como se comporta un objeto respecto a el coeficiente de elasticidad y amortiguamiento.
> #### 1.2 Objetivos específicos
> - Descubrir como se comporta un objeto para modificar la estructura del mismo antes de que la onda lo afecte para evitar fracturas.


\begin{align}
\frac{d^2x}{dt^2} &=\frac{-b}{m}*x_2 + \frac{-k}{m}*x_1 - f^2sin(2\pi*f*t)\\
\end{align}


Las propiedades del oscilador son:
- su masa m, 
- su rigidez elástica k y
- su constante de amortiguamiento b. 

Si este oscilador se somete a un acelerograma, el valor máximo de aceleración que sufrirá la masa depende de su frecuencia y su amortiguamiento. 


```python
import numpy as np
from scipy.integrate import odeint
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
# Parámetros
b = 10
m = 15
k = 200
# Modelo del sistema
def barra(x, t, f):
    x1 = x[0]
    x2 = x[1]
    return [x2, -b/m*x2-k/m*x1-f**2*np.sin(2*np.pi*f*t)]
```

```python
x0 = [0, 0]
tt = np.linspace(0,30,500)
xx = odeint(barra, x0, tt, (1,))

x1 = xx[:, 0]
x2 = xx[:, 1]
```

```python
plt.plot(tt,x1)
plt.xlabel('$t$')
plt.ylabel('Amplitud')

plt.grid()
```

```python
freq = np.logspace(-2, 2, 50)
A = []
for f in freq:
    xx = odeint(barra, x0, tt, (f,))
    x1 = xx[400:, 0]
    A.append(np.max(x1))
```

```python
plt.semilogx(freq, A)
plt.xlabel('Hertz')
plt.ylabel('Amplitud')
plt.grid()
```

```python
np.sqrt(k/m)/(2*np.pi)
```
