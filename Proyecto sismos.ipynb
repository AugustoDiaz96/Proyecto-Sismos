# Proyecto sismos 1

```python
import numpy as np
from scipy.integrate import odeint
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
# Par√°metros
b = 10
m = 15
k = 200000
# Modelo del sistema
def barra(x, t, f):
    x1 = x[0]
    x2 = x[1]
    return [x2, -b/m*x2-k/m*x1-f**2*np.sin(2*np.pi*f*t)]
```

```python
x0 = [0, 0]
tt = np.linspace(0,30,500)
xx = odeint(barra, x0, tt, (1,))

x1 = xx[:, 0]
x2 = xx[:, 1]
```

```python
plt.plot(tt,x1)
```

```python
freq = np.logspace(-2, 2, 50)
A = []
for f in freq:
    xx = odeint(barra, x0, tt, (f,))
    x1 = xx[400:, 0]
    A.append(np.max(x1))
```

```python
plt.semilogx(freq, A)
```

```python
np.sqrt(k/m)/(2*np.pi)
```
